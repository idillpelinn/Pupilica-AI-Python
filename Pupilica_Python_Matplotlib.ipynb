# **1.Line Plot**

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(6,2)) #(genislik, yukseklik)
#line plot, (x ekseni, y ekseni, renk, saydamlık, etiket)
plt.plot(veri["Id"], veri["SepalLengthCm"], color="#000fff", alpha=0.9 , label="SepalLengthCm Etiket" )
plt.plot(veri["Id"], veri["PetalLengthCm"], color="red", alpha=0.9 , label="PetalLengthCm Etiket" )

#baslik
plt.title("Iris Cicegi Gorsellestirme")

#x ekseni etiketi
plt.xlabel("Id")

#y ekseni etiketi
plt.ylabel("Sepal Length Cm ")

#izgara görünümü
plt.grid(True)

#plt plotun içerisinde yazan label görünsün diye
plt.legend()

plt.show()
```

# **2. Scatter Plot**

```python
#Species sütunundaki değerleri benzersiz olarak döndürür.
veri.Species.unique()
```

```python
veri_setosa = veri[veri.Species == 'Iris-setosa']
veri_versicolor = veri[veri.Species == 'Iris-versicolor']
veri_virginica = veri[veri.Species == 'Iris-virginica']
```

```python
plt.figure(figsize=(6,2))

#scatter(x, y, renk, size, saydamlık, label)
plt.scatter(veri_setosa["SepalLengthCm"], veri_setosa["PetalLengthCm"], color="black", s=15, alpha=0.7, label = "veri_setosa")
plt.scatter(veri_versicolor["SepalLengthCm"],veri_versicolor["PetalLengthCm"], color="red", s=15, alpha=0.7, label = "veri_versicolor")
plt.scatter(veri_virginica["SepalLengthCm"], veri_virginica["PetalLengthCm"], color="blue", s=15, alpha=0.7, label = "veri_virginica")

plt.title("İris Çiçeği Gösrelleştirme")
plt.xlabel("Sepal Length Cm")
plt.ylabel("Petal Length Cm")
plt.grid(True)
plt.legend()
plt.show()
```

# **3. Histogram**


* Boy: 150 cm ile 190 cm arasında herhangi bir değer alabilir.
* Sıcaklık: 10°C ile 30°C arasında herhangi bir değer alabilir.
* Ağırlık: 50 kg ile 100 kg arasında herhangi bir değer alabilir.4

Histogram grafiği, sürekli değişkenin değerlerini belirli aralıklara (sınıflara) böler ve her aralıktaki veri sayısını (frekansı) gösterir. Bu sayede verilerin nasıl dağıldığını, merkez eğilimini ve yayılımını daha kolay anlamak mümkün olur.

```python
import pandas as pd
veri = pd.read_csv("/content/iris.csv")
veri
```

```python
plt.figure()
plt.hist(veri["SepalLengthCm"], color="red", alpha=0.9, label="SepalLengthCm")
plt.hist(veri["PetalLengthCm"], color="blue", alpha=0.9, label="PetalLengthCm")
plt.title("Histogram")
plt.xlabel("cm")
plt.ylabel("frequency")
plt.legend()
```

# **4. Subplots**


```python
import pandas as pd
veri = pd.read_csv("/content/iris.csv")
veri

fig, ax = plt.subplots(2,1)
fig.suptitle("Subplots")
ax[0].plot(veri.SepalLengthCm, color="red")# ax[0] :ilk eksen 1.plot
ax[0].set_title("SepalLengthCm")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("cm")

ax[1].plot(veri.PetalLengthCm, color="blue")# ax[1] :ikinci eksen 1.plot
ax[1].set_title("PetalLengthCm")
ax[1].set_xlabel("Index")
ax[1].set_ylabel("cm")



fig.tight_layout()
fig.subplots_adjust(top = 0.85)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 2x2'lik bir grid içinde 4 alt grafik oluştur
plt.subplot(4,4, 1) # 2 satir ve 2 sutundan olussun ve ben 1. plotu cizdiriyorum
plt.scatter([1, 2, 3, 4], [5, 6, 7, 8])
plt.title("Alt Grafik 1")

plt.subplot(4,4, 2)
plt.plot([1, 2, 3, 4], [8, 7, 6, 5])
plt.title("Alt Grafik 2")

plt.subplot(4,4, 3) # 2 satir ve 2 sutundan olussun ve ben 3. plotu cizdiriyorum
plt.hist([1, 2, 3, 4, 5, 6, 7, 8])
plt.title("Alt Grafik 3")

plt.subplot(4,4, 4)
plt.plot([1, 2, 3, 4], [8, 7, 6, 5])
plt.title("Alt Grafik 4")

plt.show() # Grafikleri göster
```

```python
# class'larin (target variable) diagnosis ayrilmasi
# 2,3 bir subplot cizdirelim
# ilk satirda     M için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
# ikinici satirda B için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)

import pandas as pd
import matplotlib.pyplot as plt

veri = pd.read_csv("/content/breast-cancer-data.csv")
veri
```

```python
m = veri[veri["diagnosis"]=="M"]
m.reset_index(drop=True, inplace=True)

b = veri[veri["diagnosis"]=="B"]
b.reset_index(drop=True, inplace=True)

# ilk satirda     M için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
plt.subplot(2, 3, 1)
plt.scatter(m.radius_mean, m.texture_mean)

plt.subplot(2, 3, 2)
plt.hist(m.radius_mean)

plt.subplot(2, 3, 3)
plt.plot(m.radius_mean)


# ikinici satirda B için scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
plt.subplot(2, 3, 4)
plt.scatter(b.radius_mean, b.texture_mean)

plt.subplot(2, 3, 5)
plt.hist(b.radius_mean)

plt.subplot(2, 3, 6)
plt.plot(b.radius_mean)
```
